```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
D = iris.data
L = iris.target
```

```python
D
```

```python
L
```

```python
label_name = {i: name for i, name in enumerate(iris.target_names)}
label_name
```

```python
#rappresentiamo ora ogni classe separatamente per ogni caratteristica in un istogramma 
```

```python
feature_dict = {i: name for i, name in enumerate(iris.feature_names)}
feature_dict
```

```python
def generateHistograms(D,L):
    for i in feature_dict.keys():
        for j in label_name.keys():
            mask=(L==j)
            plt.hist(D[mask,i],density=True,label=label_name.get(j) )
            plt.xlabel(feature_dict.get(i))
            plt.ylabel('Density')
        plt.legend()
        plt.show()

```

```python
generateHistograms(D,L)
```

```python
#Scatter Plot tra caratteristiche
```

```python
def getScatterPlots(D,L):
    for i in feature_dict.keys():
        for j in feature_dict.keys():
            if i != j:
                for l in label_name.keys():
                    mask = (L == l)
                    plt.scatter(D[mask, i], D[mask, j], label=label_name[l])
                plt.xlabel(feature_dict[i])
                plt.ylabel(feature_dict[j])
                plt.legend()
                plt.show()
getScatterPlots(D,L)
```
