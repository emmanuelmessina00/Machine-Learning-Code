```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sklearn.datasets
def load_iris():
    D,L = sklearn.datasets.load_iris()['data'].T, sklearn.datasets.load_iris()['target']
    return D, L
D,L=load_iris()
```

```python
def split_db_2to1(D, L, seed=0):
    nTrain = int(D.shape[1]*2.0/3.0)
    np.random.seed(seed)
    idx = np.random.permutation(D.shape[1])
    idxTrain = idx[0:nTrain]
    idxTest = idx[nTrain:]
    DTR = D[:, idxTrain]
    DTE = D[:, idxTest]
    LTR = L[idxTrain]
    LTE = L[idxTest]
    return (DTR, LTR), (DTE, LTE)
    D, L = load_iris()
    # DTR and LTR are training data and labels, DTE and LTE are evaluation (or more precisely validation) data and labels
(DTR, LTR), (DTE, LTE) = split_db_2to1(D, L)
```

```python
D0=DTR[:,LTR==0]
D0 #Sono i dati con la label 0
```

```python
D1=DTR[:,LTR==1]
D2=DTR[:,LTR==2]
```

```python
def getMuAndCov(D):
    mu=D.mean(1) #in questo modo abbiamo la media delle caratteristiche
    mu=mu.reshape((mu.size, 1))
    DC=np.dot((D-mu),(D-mu).T)
    C=DC/D.shape[1]
    return mu,C
mu0,C0=getMuAndCov(D0) #vengono utilizzati i dati di training per calcolare le medie e le matrici di covarianza
mu1,C1=getMuAndCov(D1)
mu2,C2=getMuAndCov(D2)
```

```python
def logpdf_GAU_ND(X, mu, C):
    """
    Compute the log-density of a multivariate Gaussian distribution for multiple samples.

    Parameters:
    X : numpy.ndarray of shape (M, N)
        The data matrix (M features, N samples).
    mu : numpy.ndarray of shape (M, 1)
        The mean vector.
    C : numpy.ndarray of shape (M, M)
        The covariance matrix.

    Returns:
    numpy.ndarray
        The array of log-density values (N,).
    """
    M = X.shape[0]
    XC = X - mu
    invC = np.linalg.inv(C)
    log_det_C = np.linalg.slogdet(C)[1]
    term1 = -0.5 * M * np.log(2 * np.pi)
    term2 = -0.5 * log_det_C
    term3 = -0.5 * np.sum(XC * (invC @ XC), axis=0)
    return term1 + term2 + term3

ll0=logpdf_GAU_ND(DTE,mu0,C0) #Utilizzo i dati da validare
ll1=logpdf_GAU_ND(DTE,mu1,C1)
ll2=logpdf_GAU_ND(DTE,mu2,C2)

ll0,ll1,ll2
```

```python
def initSjoint(ll0,ll1,ll2):
    logS = np.vstack([ll0, ll1, ll2])  # shape (3, N)

    # Passaggio in likelihood (non strettamente necessario per classificazione, ma richiesto dall'esercizio)
    S = np.exp(logS)  # shape (3, N)

    # Calcolo delle densità con priori (uniformi)
    P_c = 1.0 / 3
    SJoint = S * P_c  # shape (3, N)
    return SJoint
Sjoint=initSjoint(ll0,ll1,ll2)
Sjoint
```

```python
def vrow(x):
	return x.reshape((1, x.size))
#Calcoliamo ora le probabilità a posteriori
def getSPost(Sjoint):
    SMarginal=vrow(Sjoint.sum(0))
    Spost=Sjoint/SMarginal
    return Spost
Spost=getSPost(Sjoint)
Spost
```

```python
predicted_labels= np.argmax(Spost,axis=0)
predicted_labels
```

```python
accuracy = (predicted_labels == LTE).mean()
print("Accuracy:", accuracy)    
```

```python
err= 1-accuracy
print("Error rate", err*100)
```

### Naive Bayes

```python
C0_nb = C0 * np.eye(C0.shape[0])
C1_nb = C1 * np.eye(C1.shape[0])
C2_nb = C2 * np.eye(C2.shape[0])

C0_nb,C1_nb,C2_nb
```

```python
ll0_nb = logpdf_GAU_ND(DTE, mu0, C0_nb)
ll1_nb = logpdf_GAU_ND(DTE, mu1, C1_nb)
ll2_nb = logpdf_GAU_ND(DTE, mu2, C2_nb)

ll0_nb, ll1_nb, ll2_nb
```

```python
logS_nb = np.vstack([ll0_nb, ll1_nb, ll2_nb])
S_nb = np.exp(logS_nb)
SJoint_nb = S_nb * (1.0 / 3)
SMarginal_nb = vrow(SJoint_nb.sum(0))
SPost_nb = SJoint_nb / SMarginal_nb
predicted_labels_nb = np.argmax(SPost_nb, axis=0)
predicted_labels_nb
```

```python
accuracy = (predicted_labels == LTE).mean()
print("Accuracy:", accuracy)
err= 1-accuracy
print("Error rate", err*100)
```

### Matrice condivisa

```python
N0 = D0.shape[1]
N1 = D1.shape[1]
N2 = D2.shape[1]
N = DTR.shape[1]

C_tied = (N0 * C0 + N1 * C1 + N2 * C2) / N
C_tied
```

```python
ll0_tied = logpdf_GAU_ND(DTE, mu0, C_tied)
ll1_tied = logpdf_GAU_ND(DTE, mu1, C_tied)
ll2_tied = logpdf_GAU_ND(DTE, mu2, C_tied)
```

```python
logS_tied = np.vstack([ll0_tied, ll1_tied, ll2_tied])
S_tied = np.exp(logS_tied)
SJoint_tied = S_tied * (1.0 / 3)
SMarginal_tied = vrow(SJoint_tied.sum(0))
SPost_tied = SJoint_tied / SMarginal_tied
predicted_labels_tied = np.argmax(SPost_tied, axis=0)
```

```python
accuracy = (predicted_labels_tied == LTE).mean()
print("Accuracy:", accuracy)
err= 1-accuracy
print("Error rate", err*100)
```

```python
mask=(L==1) | (L==2)
D_binary=D[:,mask]
L_binary=L[mask]
(DTR, LTR), (DTE, LTE) = split_db_2to1(D_binary, L_binary)
```

```python
D_B0=DTR[:,LTR==1]
D_B1=DTR[:,LTR==2]

muB0,CB0=getMuAndCov(D_B0)
muB1,CB1=getMuAndCov(D_B1)

llb1=logpdf_GAU_ND(DTE,muB0,CB0)
llb2=logpdf_GAU_ND(DTE,muB1,CB1)

llr=llb2-llb1
llr
```

```python
predicted_labels_binary = np.where(llr > 0, 2, 1)
predicted_labels_binary
```

```python
accuracy_tied = (predicted_labels_binary == LTE).mean()
print("Binary tasks accuracy:", accuracy_tied)
print("Binary tasks err:", 1-accuracy_tied)
```
