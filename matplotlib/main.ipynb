```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,5,1000)
x
```

```python
plt.plot(x,np.sin(x),'bo',markersize=5,label='seno a punti')
plt.plot(x,np.sin(x),color='r', linestyle='--',label='seno a linee tratteggiate')

plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('Grafico funzione seno')
plt.legend()
```

```python
plt.show()
```

```python
D=np.random.normal(size=1000)
x=np.linspace(-4,4,1000)
y = (1/np.sqrt(2*np.pi)) * np.exp(-0.5 * x**2)

plt.plot(x, y, color='red', linewidth=2, label='Normale Gaussiana')
plt.legend()
plt.hist(D,bins=50,density=True,ec='black')
```
