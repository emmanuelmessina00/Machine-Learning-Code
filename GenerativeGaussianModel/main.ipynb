```python
def split_db_2to1(D, L, seed=0):
    nTrain = int(D.shape[1]*2.0/3.0)
    np.random.seed(seed)
    idx = np.random.permutation(D.shape[1])
    idxTrain = idx[0:nTrain]
    idxTest = idx[nTrain:]
    DTR = D[:, idxTrain]
    DTE = D[:, idxTest]
    LTR = L[idxTrain]
    LTE = L[idxTest]
    return (DTR, LTR), (DTE, LTE)
    D, L = load_iris()
    # DTR and LTR are training data and labels, DTE and LTE are evaluation (or more precisely validation) data and labels
(DTR, LTR), (DTE, LTE) = split_db_2to1(D, L)
```

```python
D0=DTR[:,LTR==0]
D0 #Sono i dati con la label 0
```

```python
D1=DTR[:,LTR==1]
D2=DTR[:,LTR==2]
```

```python
def getMuAndCov(D):
    mu=D.mean(1) #in questo modo abbiamo la media delle caratteristiche
    mu=mu.reshape((mu.size, 1))
    DC=np.dot((D-mu),(D-mu).T)
    C=DC/D.shape[1]
    return mu,C
mu0,C0=getMuAndCov(D0) #vengono utilizzati i dati di training per calcolare le medie e le matrici di covarianza
mu1,C1=getMuAndCov(D1)
mu2,C2=getMuAndCov(D2)
```

```python
def logpdf_GAU_ND(X, mu, C):
    """
    Compute the log-density of a multivariate Gaussian distribution for multiple samples.

    Parameters:
    X : numpy.ndarray of shape (M, N)
        The data matrix (M features, N samples).
    mu : numpy.ndarray of shape (M, 1)
        The mean vector.
    C : numpy.ndarray of shape (M, M)
        The covariance matrix.

    Returns:
    numpy.ndarray
        The array of log-density values (N,).
    """
    M = X.shape[0]
    XC = X - mu
    invC = np.linalg.inv(C)
    log_det_C = np.linalg.slogdet(C)[1]
    term1 = -0.5 * M * np.log(2 * np.pi)
    term2 = -0.5 * log_det_C
    term3 = -0.5 * np.sum(XC * (invC @ XC), axis=0)
    return term1 + term2 + term3

ll0=logpdf_GAU_ND(DTE,mu0,C0) #Utilizzo i dati da validare
ll1=logpdf_GAU_ND(DTE,mu1,C1)
ll2=logpdf_GAU_ND(DTE,mu2,C2)

ll0,ll1,ll2
```

```python
def initSjoint(ll0,ll1,ll2):
    logS = np.vstack([ll0, ll1, ll2])  # shape (3, N)

    # Passaggio in likelihood (non strettamente necessario per classificazione, ma richiesto dall'esercizio)
    S = np.exp(logS)  # shape (3, N)

    # Calcolo delle densità con priori (uniformi)
    P_c = 1.0 / 3
    SJoint = S * P_c  # shape (3, N)
    return SJoint
Sjoint=initSjoint(ll0,ll1,ll2)
Sjoint
```

```python
def vrow(x):
	return x.reshape((1, x.size))
#Calcoliamo ora le probabilità a posteriori
def getSPost(Sjoint):
    SMarginal=vrow(Sjoint.sum(0))
    Spost=Sjoint/SMarginal
    return Spost
Spost=getSPost(Sjoint)
Spost
```

```python
predicted_labels= np.argmax(Spost,axis=0)
predicted_labels
```

```python
accuracy = (predicted_labels == LTE).mean()
print("Accuracy:", accuracy)    
```
