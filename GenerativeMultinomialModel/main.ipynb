```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys

def load_data():

    lInf = []

    f=open('data/inferno.txt', encoding="ISO-8859-1")

    for line in f:
        lInf.append(line.strip())
    f.close()

    lPur = []

    f=open('data/purgatorio.txt', encoding="ISO-8859-1")

    for line in f:
        lPur.append(line.strip())
    f.close()

    lPar = []

    f=open('data/paradiso.txt', encoding="ISO-8859-1")

    for line in f:
        lPar.append(line.strip())
    f.close()
    
    return lInf, lPur, lPar

def split_data(l, n):

    lTrain, lTest = [], []
    for i in range(len(l)):
        if i % n == 0:
            lTest.append(l[i])
        else:
            lTrain.append(l[i])
            
    return lTrain, lTest

if __name__ == '__main__':

    # Load the tercets and split the lists in training and test lists
    
    lInf, lPur, lPar = load_data()

    lInf_train, lInf_evaluation = split_data(lInf, 4)
    lPur_train, lPur_evaluation = split_data(lPur, 4)
    lPar_train, lPar_evaluation = split_data(lPar, 4)
    
    print(lInf_train)
```

```python
#Creazione dizionario con tutte le parole

def initDict(allTercets):
    allwords= {}
    eps= 0.001
    for t in allTercets:
        words = [w.lower() for w in t.split()]
        for w in words:
            if w not in allwords:
                allwords[w]=eps
    
    return allwords

allwords=initDict(lInf_train+lPur_train+lPar_train)
allwords
```

```python
infernowords=allwords.copy()
purgatoriowords=allwords.copy()
paradisowords=allwords.copy()
```

```python

#Per ogni cantica calcoliamo logPIc,j=log(Ncj/Nc) che andremo poi ad usare per il ML
def getLogPiPerCanticasWord(cantica,wordsOfCantica):
    
    eps=0.001
    for c in cantica:
        words=c.split()
        for w in words:
            if w in wordsOfCantica.keys():
                val=wordsOfCantica.get(w)
                val=val+1
                wordsOfCantica[w]=val
            else:
                wordsOfCantica[w]=eps
    n_words=len(wordsOfCantica.keys())
    
    for w, v in list(wordsOfCantica.items()):
        wordsOfCantica[w] = np.log(v / n_words)
        
    

getLogPiPerCanticasWord(lInf_train,infernowords)
getLogPiPerCanticasWord(lPar_train,paradisowords)
getLogPiPerCanticasWord(lPur_train,purgatoriowords)

infernowords
paradisowords
purgatoriowords
```

```python
def getlloftercets(allvalues,dictCantica):
    
   
    llf=np.zeros(len(allvalues))
    index=0
    
    for row in allvalues:
        words=row.split()
        ll=0
        
        for w in words:
            if w.lower() in dictCantica:
                ll+=dictCantica[w.lower()]
        llf[index]=ll
        index+=1
   
    return llf

llfInferno=getlloftercets(lInf_evaluation+lPar_evaluation+lPur_evaluation,infernowords)
llfParadiso=getlloftercets(lInf_evaluation+lPar_evaluation+lPur_evaluation,paradisowords)
llfPurgatorio=getlloftercets(lInf_evaluation+lPar_evaluation+lPur_evaluation,purgatoriowords)

llfParadiso,llfInferno,llfPurgatorio

```

```python
def initSjoint(llf1,llf2,llf3):
    logS = np.vstack([llf1, llf2, llf3])  # shape (3, N)

    # Passaggio in likelihood (non strettamente necessario per classificazione, ma richiesto dall'esercizio)
    S = np.exp(logS)  # shape (3, N)

    # Calcolo delle densità con priori (uniformi)
    P_c = 1.0 / 3
    SJoint = S * P_c  # shape (3, N)
    return SJoint
Sjoint=initSjoint(llfInferno,llfParadiso,llfPurgatorio)
Sjoint
```

```python
def vrow(M):
    M.reshape(1,-1)
    return M
SMarginal = vrow(Sjoint.sum(0))
SMarginal
```

```python
SPOST=Sjoint/SMarginal
SPOST
```

```python
predicted_labels= np.argmax(SPOST,axis=0)
predicted_labels
```

```python
labels={
    0:"inferno",
    1:"purgatorio",
    2:"paradiso"
}
cnt=0
for l in predicted_labels:
    print(f'predicted label for tercet {cnt} is: {labels[l]}')
    cnt+=1
```
